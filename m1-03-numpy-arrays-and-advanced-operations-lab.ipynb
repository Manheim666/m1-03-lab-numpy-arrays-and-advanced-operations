```python
import numpy as np

rng = np.random.default_rng(seed=1)

readings = rng.normal(50, 15, size = (360,4))
print(readings)
```

<h1>Task 2</h1>

```python
scaled_readings = (readings - readings.min(axis=0)) / (readings.max(axis=0) - readings.min(axis=0))
scaled_readings
```

```python
centered_readings = readings - readings.mean(axis=0)
centered_readings
```

<h1>Task 3</h1>

```python
def bool_mask(arr):
    return (arr > 80) | (arr < 20)

def outlier_info(arr):
    mask = bool_mask(arr)
    num_outliers = np.sum(mask)  
    total = arr.size                    
    return {'Number of outliers': num_outliers, 'Fraction of outliers': num_outliers / total}


info = outlier_info(readings)
print(info)


def cleaning(arr):
    arr = arr.copy()
    arr[bool_mask(readings)] = np.nan
    return arr

cleaned_reading = cleaning(readings)
cleaned_reading
```

```python
num_mask = np.sum(bool_mask(readings))
num_nan = np.isnan(cleaned_reading).sum()
print(f'Number of masks for outliers: {num_mask}\nNumber of NaN: {num_nan}')
print(num_mask == num_nan)
```

<h1>Task 4</h1>

```python
sensor_means = np.mean(readings, axis = 0)
print(sensor_means)
```

```python
sensor_stds = np.std(readings, axis = 0)
print(sensor_stds)
```

```python
time_means = np.mean(readings, axis = 1)
print(time_means)
```

```python
print(f'Number of sensor_means: {sensor_means.size}')
print(f'Number of sensor_stds: {sensor_stds.size}')
print(f'Number of time_means: {time_means.size}')
```

<h1>Task 5<h1>

```python
report = {
    'total_readings': readings.size,
    'outlier_count': np.sum(bool_mask(readings)),
    'outlier_percent': np.sum(bool_mask(readings)) / readings.size * 100,
    'sensor_means': np.mean(readings, axis=0),
    'sensor_stds': np.std(readings, axis=0)
}

if report['total_readings'] != readings.size:
    print("Warning: total readings does not match array size!")

report_str = (
    f"Total readings: {report['total_readings']}\n"
    f"Outlier count: {report['outlier_count']}\n"
    f"Outlier percent: {report['outlier_percent']:.2f}%\n"
    f"Sensor means: {report['sensor_means']}\n"
    f"Sensor stds: {report['sensor_stds']}"
)

print(report_str)
```

<h1>Task 6<h1>

```python
import numpy as np

rng = np.random.default_rng(seed=1)

sim = rng.normal(0, 1, size = (1000,6))
print(readings)
```

```python
print(sim.shape)
print(sim.dtype)
print(sim[:3,:])
```

<h1>Task 7</h1>

```python
scenario_shift = [-0.3, -0.1, 0.0, 0.1, 0.2, 0.4]
```

```python
adjusted_sim = sim + scenario_shift
print(adjusted_sim)
```

```python
print(f'Shape of original sim: {sim.shape}')
print(f'Shape of adjusted sim: {adjusted_sim.shape}')
print(sim.shape == adjusted_sim.shape)
```

<h1>Task 8<h1>

```python
scenario_means = adjusted_sim.mean(axis=0) 
print("Scenario means:", scenario_means)

ranking = np.argsort(scenario_means)
print("Ranking of scenarios (lowest to highest):", ranking)

top2_partition = np.partition(scenario_means, -2)[-2:]
print("Top 2 scenario means (partition):", top2_partition)

top2_sorted = np.sort(scenario_means)[-2:]
print("Top 2 scenario means (full sort):", top2_sorted)

print('Matching: ',np.all(np.isin(top2_partition, top2_sorted)))
```

<h1>Task 9<h1>

```python
rng_same = np.random.default_rng(seed=1)
sim_2 = rng_same.normal(0, 1, size=(1000, 6))

print("sim and sim_2 are equal:", np.array_equal(sim, sim_2))
print("Number of matching elements:", np.sum(sim == sim_2))  
```

```python
rng_diff = np.random.default_rng(seed=2)
sim_3 = rng_diff.normal(0, 1, size=(1000, 6))

print("sim and sim_3 are approximately equal (np.allclose):", np.allclose(sim, sim_3))
print("Number of matching elements with different seed:", np.sum(sim == sim_3))
```

<h1>Task 10</h1>

```python
top_two_indices = np.argsort(scenario_means)[-2:]  
top_two_values = scenario_means[top_two_indices] 

validation_report = {
    'shape': adjusted_sim.shape,
    'scenario_means': scenario_means,
    'top_two_indices': top_two_indices,
    'top_two_values': top_two_values
}

assert len(scenario_means) == 6, "scenario_means should have length 6"
assert len(top_two_indices) == 2, "top_two_indices should have length 2"

report_str = (
    f"Shape of adjusted_sim: {validation_report['shape']}\n"
    f"Scenario means: {validation_report['scenario_means']}\n"
    f"Indices of top 2 scenarios: {validation_report['top_two_indices']}\n"
    f"Values of top 2 scenarios: {validation_report['top_two_values']}"
)

print(report_str)
```
